```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
from matplotlib import pyplot
from pylab import genfromtxt
switch1 = genfromtxt("switch1.txt")
switch2 = genfromtxt("switch2.txt")
s1adj = genfromtxt("switch123.txt")
s2adj = genfromtxt("switch223.txt")
VS1 = q.MeasurementArray(switch1[:36,0],error=[0.05],name="Voltage",units="V")
CS1 = q.MeasurementArray(switch1[:36,1],error=[0.005],name="Current",units="mA")
VS2 = q.MeasurementArray(switch2[:43,0],error=[0.05],name="Voltage",units="V")
CS2 = q.MeasurementArray(switch2[:43,1],error=[0.005],name="Current",units="mA")
VS1adj = q.MeasurementArray(s1adj[:28,0],error=[0.05],name="Voltage",units="V")
CS1adj = q.MeasurementArray(s1adj[:28,1],error=[0.005],name="Current",units="mA^(2/3)")
VS2adj = q.MeasurementArray(s2adj[:32,0],error=[0.05],name="Voltage",units="V")
CS2adj = q.MeasurementArray(s2adj[:32,1],error=[0.005],name="Current",units="mA^(2/3)")
```

#### Part 1: Switches 

```python
sw1fig = q.MakePlot(VS1,CS1)
sw2fig = q.MakePlot(VS2,CS2)
s123fig = q.MakePlot(VS1adj,CS1adj)
s223fig = q.MakePlot(VS2adj,CS2adj)
sw1fig.show()
sw2fig.show()
```

```python
def linFit(x, *para):
    return para[0]*x + para[1]
```

```python
s123fig.fit(linFit,fitcolor="blue",parguess=[1,1])
s223fig.fit(linFit,fitcolor="blue",parguess=[1,1])
s123fig.show()
s223fig.show()
```
