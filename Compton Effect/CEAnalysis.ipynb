```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
```

```python
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
CE60 = genfromtxt("CE60calibrated.txt")
CE90 = genfromtxt("CE90calibrated.txt")
CE120= genfromtxt("CE120calibrated.txt")
```

```python
enA = q.MeasurementArray(CE60[:500,0],error=[2],name="Energy A",units="keV")
enB = q.MeasurementArray(CE60[:500,2],error=[2],name="Energy B",units="keV")
count60A = q.MeasurementArray(CE60[:500,1],error=[1],name="Counts A",units = "n")
count60B = q.MeasurementArray(CE60[:500,3],error=[1],name="Counts B",units = "n")
```

```python
def gaussianFit(x, *para):
    return para[0]*np.exp(-((x-para[1])**2)/(2*para[2]**2))+para[3]*x+para[4]
```

```python
figure1 = q.MakePlot(enA,count60A)
figure2 = q.MakePlot(enB,count60B)
figure1.fit(gaussianFit,fitcolor="blue",parguess=[40,220,15,0,0])
figure2.fit(gaussianFit,fitcolor="blue",parguess=[40,400,15,0,0])
figure1.show()
figure2.show()
```

###### 60 degree results:
Scintillator A energy = 234.6 +/- 48.0 keV<br />
Scintillator B energy = 406.0 +/- 43.0 keV 

```python
count90A = q.MeasurementArray(CE90[:500,1],error=[1],name="Counts A",units = "n")
count90B = q.MeasurementArray(CE90[:500,3],error=[1],name="Counts B",units = "n")
figure3 = q.MakePlot(enA,count90A)
figure4 = q.MakePlot(enB,count90B)
figure3.fit(gaussianFit,fitcolor="blue",parguess=[60,300,15,0,0])
figure4.fit(gaussianFit,fitcolor="blue",parguess=[60,290,15,0,0])
figure3.show()
figure4.show()
```

###### 90 degree results:
Scintillator A energy = 344.7 +/- 38.8 keV<br />
SCintillator B energy = 283.6 +/- 30.6 keV

```python
count120A = q.MeasurementArray(CE120[:500,1],error=[1],name="Counts A",units = "n")
count120B = q.MeasurementArray(CE120[:500,3],error=[1],name="Counts B",units = "n")
figure5 = q.MakePlot(enA,count120A)
figure6 = q.MakePlot(enB,count120B)
figure5.fit(gaussianFit,fitcolor="blue",parguess=[60,300,15,0,0])
figure6.fit(gaussianFit,fitcolor="blue",parguess=[60,290,15,0,0])
figure5.show()
figure6.show()
```
