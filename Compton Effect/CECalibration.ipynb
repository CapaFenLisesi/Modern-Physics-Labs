```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
```

```python
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
cal22Na = genfromtxt("CEcalibrationNa.txt")
cal133Ba = genfromtxt("CEcalibrationBa.txt")
cal137Cs = genfromtxt("CEcalibrationCs.txt")
```

```python
bins = q.MeasurementArray(cal137Cs[:500,0],error=[0], name="Bin",units="n")
countCsA = q.MeasurementArray(cal137Cs[:500,1],error=[2], name="Count of Scintillator A for 137Cs",units="n")
countCsB = q.MeasurementArray(cal137Cs[:500,2],error=[2], name="Count of Scintillator B for 137Cs",units="n")
countBaA = q.MeasurementArray(cal133Ba[:500,1],error=[2], name="Count of Scintillator A for 133Ba",units="n")
countBaB = q.MeasurementArray(cal133Ba[:500,2],error=[2], name="Count of Scintillator B for 133Ba",units="n")
countNaA = q.MeasurementArray(cal22Na[:500,1],error=[1], name="Count of Scintillator A for 22Na",units="n")
countNaB = q.MeasurementArray(cal22Na[:500,2],error=[1], name="Count of Scintillator B for 22Na",units="n")
```

# 137Cs Calibration

```python
figure1 = q.MakePlot(bins,countCsA)
figure2 = q.MakePlot(bins,countCsB)
figure1.show()
figure2.show()
```

```python
binsA = q.MeasurementArray(cal137Cs[250:400,0],error=[0], name="Bin",units="n")
binsB = q.MeasurementArray(cal137Cs[250:400,0],error=[0], name="Bin",units="n")
countCsASum = q.MeasurementArray(cal137Cs[250:400,1],error=[1], name="Count of Scintillator A for 137Cs",units="n")
countCsBSum = q.MeasurementArray(cal137Cs[250:400,2],error=[1], name="COunt of Scintillator B for 137Cs",units="n")
```

```python
figure3 = q.MakePlot(binsA,countCsASum)
figure4 = q.MakePlot(binsB,countCsBSum)
```

```python
def gaussianFit(x, *para):
    return para[0]*np.exp(-((x-para[1])**2)/(2*para[2]**2))+para[3]*x+para[4]
```

```python
figure3.fit(gaussianFit,fitcolor="blue",parguess=[80,600,30,1,1])
figure4.fit(gaussianFit,fitcolor="blue",parguess=[80,600,30,1,1])
figure3.show()
figure4.show()
```

Assuming the peak of the 137Cs calibration is located at 662 keV. <br /> <br />
The Scintillator A peak was located at: 590.6 with an std of 27.2 <br /> <br />
The Scintillator B peak was located at: 614.9 with an std of 23.8

# Linear Calibration

```python
binLinA = q.MeasurementArray([81.8,472,590.6],error=[6,24,27.2],name="Bin A",units="n")
binLinB = q.MeasurementArray([75.8,472,614.9],error=[3.8,19,23.8],name="Bin B",units="n")
En = q.MeasurementArray([81,511,662],error=[3],name="Energy",units="keV")
figure13 = q.MakePlot(binLinA,En)
figure14 = q.MakePlot(binLinB,En)
figure13.show()
figure14,show()
```

```python
def linFit(x, *para):
    return para[0]*x+para[1]
```

```python
figure13.fit(linFit,fitcolor="blue",parguess=[1,0])
figure14.fit(linFit,fitcolor="blue",parguess=[1,0])
figure13.show()
figure14.show()
```
