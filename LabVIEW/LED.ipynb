```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
cal = genfromtxt("SamNickOpenVoltage.txt")
PMWcal = q.MeasurementArray(cal[0,1:256],error=[0],name="PMW input",units="")
Vcal = q.MeasurementArray(cal[1,1:256],error=[0.05],name="Voltage",units="V")
```

```python
figure = q.MakePlot(PMWcal,Vcal)
figure.show()
```

```python
def linFit(x, *para):
    return para[0]*x + para[1]
```

```python
figure.fit(linFit,fitcolor="blue",parguess=[0.01,0.1193])
figure.show()
```

```python
blue = genfromtxt("Blue.txt")
blueV = q.MeasurementArray(blue[:,0],error=[0.05],name="Input Voltage",units="V")
blueI = q.MeasurementArray(blue[:,1],error=[0.005],name="Photodiode Current",units="mA")
figure1 = q.MakePlot(blueV,blueI)
figure1.show()
```

```python
bfV = q.MeasurementArray(blue[150:256,0],error=[0.05],name="Input Voltage",units="V")
bfI = q.MeasurementArray(blue[150:256,1],error=[0.005],name="Photodiode Current",units="mA")
bluefit = q.MakePlot(bfV,bfI)
bluefit.fit(linFit, fitcolor="blue", parguess=[0.001,0])
bluefit.show()
```
