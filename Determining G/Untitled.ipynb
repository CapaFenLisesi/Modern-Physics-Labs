```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
```

```python
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
mat0 = genfromtxt("300secTrial1.txt");
```

```python
time = q.MeasurementArray(mat0[:10000,0], error=[0],name="time", units="seconds")
```

```python
disp = q.MeasurementArray(mat0[:10000,1], error=[1],name="displacement", units="counts")
```

```python
figure = q.MakePlot(time,disp)
figure.show()
```

```python
def pendulumModel(x, *para):
    return para[0]*np.sin(para[1]*x + para[2])*np.exp(-x*para[3]) + para[4]
```

```python
figure.fit(pendulumModel, parguess=  [50,4,1,0.0001,-45]);
```

```python
figure.show()
```

```python
residual = q.MeasurementArray(figure.datasets[0].fit_yres[0],error=[0],name="displacement residual",units="counts")
```

```python
residualfig = q.MakePlot(time,residual)
residualfig.show()
```
