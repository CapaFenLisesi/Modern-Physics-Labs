```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
```

```python
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
muonData = genfromtxt("MuonSimpleData.txt")
```

```python
print(muonData)
```

```python
muonCount = q.MeasurementArray(muonData,error=[1],name="Muon Count", units="N")
```

```python
bins = q.MeasurementArray(np.linspace(1,len(muonData),len(muonData)),error=[1],name="Bins",units="N")
```

```python
figure = q.MakePlot(bins,muonCount)
figure.show()
```
