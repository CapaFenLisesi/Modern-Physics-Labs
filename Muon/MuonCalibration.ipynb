```python
import qexpy as q
import numpy as np
```

```python
q.plot_engine = "mpl"
```

```python
from matplotlib import pyplot;
from pylab import genfromtxt;
```

```python
muonCalData = genfromtxt("MuonSimpleCal.txt")
```

```python
muonCalData = muonCalData[:1000]
```

```python
muonCalCount = q.MeasurementArray(muonCalData,error=[100],name="Muon Count",units="n")
```

```python
bins = q.MeasurementArray(np.linspace(1,len(muonCalData),len(muonCalData)),error=[1],name="Bin",units="n")
```

```python
figure = q.MakePlot(bins,muonCalCount)
figure.show()
```

```python
def muonCal(x, *para):
    return para[0]*np.exp(-((x-para[1])**2)/(2*para[2]**2)) + para[3]*np.exp(-((x-para[4])**2)/(2*para[5]**2)) + para[6]*np.exp(-((x-para[7])**2)/(2*para[8]**2)) + para[9]*np.exp(-((x-para[10])**2)/(2*para[11]**2)) + para[12]*np.exp(-((x-para[13])**2)/(2*para[14]**2)) + para[15]*np.exp(-((x-para[16])**2)/(2*para[17]**2))  + para[18]*np.exp(-((x-para[19])**2)/(2*para[20]**2)) + para[21]*np.exp(-((x-para[22])**2)/(2*para[23]**2)) + para[24]*np.exp(-((x-para[25])**2)/(2*para[26]**2)) + para[27]*np.exp(-((x-para[28])**2)/(2*para[29]**2))
```

```python
figure.fit(muonCal, fitcolor="blue", parguess = [500000,101,2,2000000,165,2,2000000,267,2,1300000,359,2,1000000,441,2,1000000,524,2,1000000,623,2,700000,720,2,700000,811,2,500000,902,2])
```

```python
figure.show()
```

```python
BinCal = q.MeasurementArray([101,166,278,359,441,524,624,721,812,903],error=[1],name="Bin",units="n")
```

```python
period = q.MeasurementArray([1.99,4,6,8,10,11.99,13.96,16.04,18,19.96],error=[0.05],name="Period",units="s")
```

```python
figure = q.MakePlot(BinCal,period)
figure.show()
```

```python
def xaxisCal(n, *para):
    return para[0]*n+para[1]
```

```python
figure.fit(xaxisCal, parguess=[0.004,0])
```
